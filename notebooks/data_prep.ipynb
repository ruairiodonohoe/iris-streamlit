```python
from sklearn import datasets
import pandas as pd
import re
```

```python
iris = datasets.load_iris(as_frame=True)
df = iris.frame
df['species'] = df.target.map(lambda x: iris.target_names[x])
```

```python
col_names = [re.sub("[()]", "", col.replace(" ", "_"), ) for col in df.columns]
df.columns = col_names
```

```python
print(iris.filename)
df.to_csv(f"../resources/data/{iris.filename}", index=False)
```
